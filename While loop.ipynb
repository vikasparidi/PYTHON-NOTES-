```python
while a<=b:             ### checking condition of a is less or equal to b
    print(a)            ### print the value of a
    a=a+1               ### intial a=1 , now a=1+1,a=2 and checks the next condition 
```

# Difference between FOR and WHILE loop:

In for loop once the condition is satisfied it will enter the loop at one time and doesnot go back to the same statement to chect the condition.

Where as in while loop the condition will go on checking, by entering the loop again and again till the condition is true.

```python
###Point to be remember: We can also use else in while same like that we use in for loop.
```

```python
a=1
b=10
```

```python
while a<=b:
    print(a)
    a=a+2
else:                                          ### on the successful completion of while loop we can see execution of else block,
    print("print this else block ")
```

```python
a=1
b=10
while a<10:
    print(a)
    if a == 5:
        break              ### if we apply break in between else doesnot perform 
    a=a+1
else:
    print("print this")
```

```python
### point to be remember: if we break any loop(while or for) in between else block doesn't perform
```

```python
a=1
b=10
while a<b:
    print(a)
    if a == 3:                  ### continue statement doen't execute what is been written after that
        continue                ### Because of continue statement we can get the infinite loop 
    a=a+1
```

```python
a=1
b=10
while a<b:
    a=a+1
    if a==3:
        continue      ###gives control immediately to while loop, so 3 is not printed.
    print(a)
```

```python
a=1
b=10
while a<b:
    a=a+1
    if a == 3:
        pass
    print(a)
```
