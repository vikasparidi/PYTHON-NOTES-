# Difference between FOR and WHILE loop:

In for loop once the condition is satisfied it will enter the loop at one time and doesnot go back to the same statement to chect the condition.

Where as in while loop the condition will go on checking, by entering the loop again and again till the condition is true.

QUESTION 11:WRITE A FUNCTION THAT CONVERTS ENTIRE DICTIONARY VALUES TO LIST EVEN IN CASE OF NESTED DIC 

```python
d={"k1":"v1","k2":"v2","k3":{"k12":"v12","k13":"v13","k14":"v14"}}
```

```python
def dic_list(d):
    l1=[]
    if type(d) == dict:
        for i in d.values():
            if type(i)!= dict:
                l1.append(i)
            else:
                for j in i.values():
                    l1.append(j)
    return l1

            
    
             ###or
        
    l=[]
    if type(d) == dict:
        for i in d.value():
            if type(i)!= dict:
                l.append(i)
            if type(i) == dict:
                l.append(i)
        return l
```

```python
dic_list({"k1":"v1","k2":"v2","k3":{"k12":"v12","k13":"v13","k14":"v14"}})
```
